```python
from bs4 import BeautifulSoup as bs
```

```python
import urllib.request
```

```python
url_read = urllib.request.urlopen('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').read()
to_bs = bs(url_read, 'html')
```

```python
table = to_bs.find('table')
```

```python
table_rows = table.find_all('tr')
```

```python
new_list = []
for tr in table_rows:
    td = tr.find_all('td')
    for i in td:
        if '\n' in i:
            i.replace('\n', ' ')
        new_list.append(i.text)
for i in new_list:
    if '\n' in i:
        i.replace('\n', ' ')
new_list
```
